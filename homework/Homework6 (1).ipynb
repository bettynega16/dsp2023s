```python
import pandas as pd

# Read the data from wdbc.data into a pandas DataFrame
data = pd.read_csv('wdbc.data', header=None)

# Set column names based on the wdbc.names file
column_names = ['id', 'diagnosis', 'radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean', 'smoothness_mean',
                'compactness_mean', 'concavity_mean', 'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean',
                'radius_se', 'texture_se', 'perimeter_se', 'area_se', 'smoothness_se', 'compactness_se',
                'concavity_se', 'concave points_se', 'symmetry_se', 'fractal_dimension_se', 'radius_worst',
                'texture_worst', 'perimeter_worst', 'area_worst', 'smoothness_worst', 'compactness_worst',
                'concavity_worst', 'concave points_worst', 'symmetry_worst', 'fractal_dimension_worst']
data.columns = column_names

# Display the header of the data using pandas
print(data.head())

# Remove the 'id' column since it is not useful for analysis
data = data.drop('id', axis=1)

# Convert the 'diagnosis' column to numerical values
data['diagnosis'] = data['diagnosis'].map({'M': 1, 'B': 0})

# Extract the column of labels from the data frame and store it as a variable called labels
labels = data['diagnosis']

# Inspect the modified data frame
print(data.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data into a pandas dataframe
data = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/wdbc.data', header=None)

# Assign column names to the dataframe
data.columns = ['id', 'diagnosis', 'radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean', 'smoothness_mean',
                'compactness_mean', 'concavity_mean', 'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean',
                'radius_se', 'texture_se', 'perimeter_se', 'area_se', 'smoothness_se', 'compactness_se',
                'concavity_se', 'concave points_se', 'symmetry_se', 'fractal_dimension_se', 'radius_worst',
                'texture_worst', 'perimeter_worst', 'area_worst', 'smoothness_worst', 'compactness_worst',
                'concavity_worst', 'concave points_worst', 'symmetry_worst', 'fractal_dimension_worst']

# Convert the diagnosis column to binary
data['diagnosis'] = data['diagnosis'].map({'M':1, 'B':0})

# 1. Plot the data in 5 different ways of your choosing

# Scatter plot of radius_mean vs texture_mean, with color representing diagnosis
sns.scatterplot(x='radius_mean', y='texture_mean', hue='diagnosis', data=data)
plt.xlabel('Radius Mean')
plt.ylabel('Texture Mean')
plt.show()

# Histogram of area_mean, separated by diagnosis
sns.histplot(x='area_mean', hue='diagnosis', data=data, kde=True)
plt.xlabel('Area Mean')
plt.show()

# Boxplot of radius_worst, separated by diagnosis
sns.boxplot(x='diagnosis', y='radius_worst', data=data)
plt.xlabel('Diagnosis')
plt.ylabel('Radius Worst')
plt.show()

# Violinplot of perimeter_mean, separated by diagnosis
sns.violinplot(x='diagnosis', y='perimeter_mean', data=data)
plt.xlabel('Diagnosis')
plt.ylabel('Perimeter Mean')
plt.show()

# Heatmap of correlation matrix
corr_matrix = data.drop(['id', 'diagnosis'], axis=1).corr()
sns.heatmap(corr_matrix, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
